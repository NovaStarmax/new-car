```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
x, y = make_regression(n_samples=100, n_features=1, noise=1)
y = y.reshape(-1,1)
plt.scatter(x,y)
```

```python
X = np.hstack((x, np.ones(x.shape)))
X.shape
```

```python
theta = np.random.randn(2,1)
theta.shape
```

```python
def model(x, theta):
    return x @ theta
```

```python
plt.scatter(x,y)
plt.plot(x, model(X, theta), c='r')
```

```python
def thetha_analytique(X, y):
    return np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
optimisation = X @ thetha_analytique(X, y)
```

```python
plt.scatter(x, y)
plt.plot(x, optimisation, c='g', label='Solution analytique')
plt.legend()
plt.show()
```
